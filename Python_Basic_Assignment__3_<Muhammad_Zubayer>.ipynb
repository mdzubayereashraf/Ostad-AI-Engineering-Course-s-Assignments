#**Ans to the question no.04**






```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7}

# Finding the common elements in both sets
common_elements = set1.intersection(set2)
print(common_elements)

# Finding the unique elements in both sets
unique_elements = set1.union(set2)
print(unique_elements)

```
